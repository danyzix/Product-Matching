```python
# import2 - product specs
df_specs = pd.read_csv('/home/danyzix/Data/Flix_products/product_specifications.csv')
# count the number if data in each columns
df_specs.apply(lambda x: x.count(), axis=0)
```

```python
df_specs.head()
```

```python
df_specs['title:value'] = df_specs['title'] + ':' + df_specs['value']
```

```python
df_specs.head()
```

```python
# replace NA by empty string
specs_list = df_specs.groupby('product_id')['title:value'].unique().reset_index()
specs_list.fillna('', inplace=True)
specs_list.head(10)
```

```python
# import3 - product titles and categories
df_titles = pd.read_csv('/home/danyzix/Data/Flix_products/product_title.csv', engine='python')
df_titles.head()
```

```python
# import4 - product titles, mpns, brands
df_mpns = pd.read_csv('/home/danyzix/Data/Flix_products/products.csv', sep='\t')
columns = ['product_id','product_title','description','manufacturer_id','brand_title']
df_mpns = df_mpns[columns]
df_mpns.head(2)
```

```python
# import5 - product keywords and mpns
df_keyword = pd.read_csv('/home/danyzix/Data/Flix_products/product_mpns.csv')
df_keyword.head()
```

```python
# replace all \N with empty strings
df_keyword = df_keyword.replace(r'\\N', np.nan, regex=True)
df_keyword.fillna('', inplace=True)
df_keyword.head()
```

```python
# extract all unique MPNs for each product
mpn_list = df_keyword.groupby('product_id')['mpn'].unique().reset_index()
mpn_list.fillna('', inplace=True)
mpn_list[:10]
```

```python
# extract all unique keywords for each product
keywords = df_keyword.groupby('product_id')['keyword'].unique()
```

```python
len(set(df_keyword['product_id']))
```

```python
# merge import 3, 4 and 5 on key fields
df_products = pd.merge(df_mpns, df_titles, on='product_id', how='inner')
df_products = pd.merge(df_products,mpn_list, on='product_id', how='left')
df_products = pd.merge(df_products,specs_list, on='product_id', how='left')

# Remove the pipes '|' in the Categories
# df_products['Categories'] = df_products['Categories'].str.replace('|', '')

# create only one title per product as there are some duplicates
df_products['new_title']=np.nan
for i in range(len(df_products)):
    if pd.isnull(df_products.loc[i, 'product_title']):
        df_products.loc[i, 'new_title'] =  df_products.loc[i, 'title'] 
    else:
        df_products.loc[i,'new_title'] =  df_products.loc[i, 'product_title']
# drop redundant columns
df_products.drop(['product_title','title'], axis=1, inplace=True)

df_products = df_products.rename(columns={'brand_title':'brand','new_title':'title','title:value':'specs'})

cols = ['title','mpn','brand','specs','Categories','description','product_id']
df_products = df_products[cols]
df_products.head(2)
```

```python
# save dataframe to disk
df_products.to_csv('/home/danyzix/Data/df_products.csv', sep='\t')
```

```python
# to load the data back
# df_products_1 = pd.read_csv('/home/danyzix/Data/df_products.csv', sep='\t')
```

```python
df_products_1 = df_products.copy()
# necessary to replace NaN with empty string for the following steps
df_products_1 = df_products_1.replace(np.nan, '', regex=True)
```

```python
# function to convert an narray to list of string
def array_to_list_column(column):
    x = []
    for i in range(len(df_products_1)):
        if len(df_products_1.loc[i,column]) > 0:   # takes care of rows which are blank
            temp = df_products_1.loc[i,column].astype(str).tolist()   # some data are blank, converting to string to list them
            temp = " ".join(temp)
        else:
            temp = ""
        x.append(temp)           
    return x
```

```python
# convert arrays to strings
df_products_1['mpns'] = array_to_list_column('mpn')
df_products_1['specifications'] = array_to_list_column('specs')
```

```python
# reorder the dataframe
col = ['title','brand','mpns','Categories','specifications','description','product_id']
df_products_1 = df_products_1[col]
df_products_1.head()
```

```python
# save dataframe as a record json format
df_products_1.to_json('/home/danyzix/Data/brand_products_json.txt', orient='records')
```

```python
# open the text data as a json object
with open('/home/danyzix/Data/brand_products_json.txt') as json_file:  
    dict_products = json.load(json_file)
json_file.close()
```
